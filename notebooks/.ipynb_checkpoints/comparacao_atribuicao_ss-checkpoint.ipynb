# Comparação entre métodos de atribuição de estruturas secundárias

```python
from glob import glob
from os.path import basename
import pandas as pd
import seaborn as sns; sns.set()
import numpy as np
```

```python
%matplotlib inline
```

```python
def get_seq(fn):
    f = open(DATA_PATH+"/seq/"+fn)
    _ = f.readline()
    seq = f.readline()
    f.close()
    return seq
    
```

```python
def get_dssp(fn):
    f = open(DATA_PATH+"/dssp/"+fn)
    _ = f.readline()
    dssp = f.readline()
    f.close()
    return dssp
    
```

```python
def get_stride(fn):
    f = open(DATA_PATH+"/stride/"+fn)
    _ = f.readline()
    stride = f.readline()
    f.close()
    return stride
    
```

```python
def get_kaksi(fn):
    f = open(DATA_PATH+"/kaksi/"+fn)
    _ = f.readline()
    kaksi = f.readline()
    f.close()
    return kaksi
    
```

```python
def get_pross(fn):
    f = open(DATA_PATH+"/pross/"+fn)
    _ = f.readline()
    pross = f.readline()
    f.close()
    return pross
    
```

```python
def create_combinations():
    algs = ["DSSP", "STRIDE", "PROSS", "KAKSI"]
    ss = ["H","E","C"]
    combs = {}
    for a1 in range(len(algs)):
        for a2 in range(a1+1,len(algs)):
            for ss1 in ss:
                for ss2 in ss:
                    combs[(algs[a1],algs[a2],ss1,ss2)] = 0
    return combs                    
```

```python
def is_val_ss(ss):
    return ss in ["H","E","C"]
```

```python
def ss1_vs_ss2(dic, ss1_name, ss1, ss2_name, ss2):
    for i in range(len(ss1)):
        if is_val_ss(ss1[i]) and is_val_ss(ss2[i]):
            dic[(ss1_name, ss2_name, ss1[i], ss2[i])] += 1
        
```

```python
def is_H(s): return s == 'H'
def is_E(s): return s == 'E'
def is_C(s): return s == 'C'
def not_def(s): return s == '?'

def analyze_consensus(c, dssp, stride, kaksi, pross):
    for i in range(len(dssp)):
        if is_H(dssp[i]) and is_H(stride[i]) and is_H(kaksi[i]) and is_H(pross[i]):
            c["H"] += 1
        elif is_E(dssp[i]) and is_E(stride[i]) and is_E(kaksi[i]) and is_E(pross[i]):
            c["E"] += 1
        elif is_C(dssp[i]) and is_C(stride[i]) and is_C(kaksi[i]) and is_C(pross[i]):
            c["C"] += 1
        elif not_def(dssp[i]) or not_def(stride[i]) or not_def(kaksi[i]) or not_def(pross[i]):
            c["NOT_DEF"] += 1 
            if not_def(dssp[i]) and not_def(stride[i]) and not_def(pross[i]):
                c["NOT_DEF_&_NOT_VIS"] += 1
        else:
            c["NOT_EQUAL"] += 1
        
```

```python
def normalize_comp(comp):
    total = 0
    comp_norm = comp.copy()
    for k in comp.keys():
        if k[0] == 'DSSP' and k[1] == 'STRIDE':
            total += comp[k]
    total
    for k in comp.keys():
        comp_norm[k] /= float(total) 
    return comp_norm
       
```

```python
DATA_PATH="/home/jgcarvalho/zeca-analyse-pos_quali/Top8000-best_hom50_pdb_chain/cba_times_mcc/rose_special_charged/run_10000"
comparative = create_combinations()
consensus = {"H":0, "E":0, "C":0, "NOT_EQUAL":0, "NOT_DEF":0, "NOT_DEF_&_NOT_VIS":0}
for fn in glob(DATA_PATH+"/seq/*"):
    f = basename(fn)
    dssp = get_dssp(f)
    stride = get_stride(f)
    kaksi = get_kaksi(f)
    pross = get_pross(f)
    ss1_vs_ss2(comparative,"DSSP", dssp, "STRIDE", stride)
    ss1_vs_ss2(comparative,"DSSP", dssp, "PROSS", pross)
    ss1_vs_ss2(comparative,"DSSP", dssp, "KAKSI", kaksi)
    ss1_vs_ss2(comparative,"STRIDE", stride, "PROSS", pross)
    ss1_vs_ss2(comparative,"STRIDE", stride, "KAKSI", kaksi)
    ss1_vs_ss2(comparative,"PROSS", pross, "KAKSI", kaksi)
    analyze_consensus(consensus, dssp, stride, kaksi, pross)
    
comparative_norm = normalize_comp(comparative)
```

```python
print(comparative)
print(comparative_norm)
print(consensus)
```

# Resultados

```python
def resultado_01(c):
    total_res = c['H'] + c['E'] + c['C'] + c['NOT_EQUAL'] + c['NOT_DEF']
    return c['NOT_DEF']/total_res
print(resultado_01(consensus))
print("{:.2f}% dos resíduos falharam ao ter a estrutura secundária atribuída por um ou mais métodos".format(resultado_01(consensus)*100.0))
    
```

```python
def resultado_02(c):
    total_res = c['H'] + c['E'] + c['C'] + c['NOT_EQUAL'] + c['NOT_DEF']
    return c['NOT_DEF_&_NOT_VIS']/total_res
print(resultado_02(consensus))
print("{:.2f}% dos resíduos falharam ao ter a estrutura secundária atribuída por todos os métodos".format(resultado_02(consensus)*100.0))
    
```

```python
def resultado_03(c):
    total_def = c['H'] + c['E'] + c['C'] + c['NOT_EQUAL'] 
    return (total_def - c['NOT_EQUAL'])/total_def
print(resultado_03(consensus))
print("Apenas {:.2f}% dos resíduos (com estrutura definida) apresentaram consenso entre os 4 métodos de atribuição de estruturas secundárias".format(resultado_03(consensus)*100.0))
    
```

```python
def resultado_04(c):
    total_def = (c['H'] + c['E'] + c['C'] + c['NOT_EQUAL']) * 4
    return ((total_def - c['NOT_EQUAL']*2)/total_def,(total_def - c['NOT_EQUAL'])/total_def)
print(resultado_04(consensus))
print("Apenas {:.2f}% {:.2f}% os resíduos (com estrutura definida) apresentaram consenso entre os 4 métodos de atribuição de estruturas secundárias".format(resultado_04(consensus)[0]*100.0,resultado_04(consensus)[1]*100.0))
    
```

```python
def plot_algs(c):
    result = np.eye(4)
    for ss in ['H','E','C']:
        result[0,1] += c[('DSSP','STRIDE',ss,ss)]
        result[0,2] += c[('DSSP','PROSS',ss,ss)]
        result[0,3] += c[('DSSP','KAKSI',ss,ss)]
        result[1,2] += c[('STRIDE','PROSS',ss,ss)]
        result[1,3] += c[('STRIDE','KAKSI',ss,ss)]
        result[2,3] += c[('PROSS','KAKSI',ss,ss)]
    labels = ["DSSP","STRIDE","PROSS","KAKSI"]
    sns.heatmap(result, annot=True,xticklabels=labels, yticklabels=labels, mask= np.tri(4, 4, -1, dtype=bool)
)
    return result
print(plot_algs(comparative_norm))
```

```python
def plot_vs(c, ss1_name, ss2_name):
    ss = ['H','E','C'] 
    result = np.zeros((3,3))
    for a in range(3):
        for b in range(3):
            result[a,b] = c[(ss1_name,ss2_name,ss[a],ss[b])]
            
    sns.heatmap(result, annot=True)
    return result    
```
